```python
import pandas as pd
```

```python
# Task 1: Load Data
candidates = pd.read_csv("candidates.csv")
constituencies = pd.read_csv('constituencies.csv')
election_results = pd.read_csv('election_results.csv')
```

```python
# Task 2: Basic Inspection
election_results.head(5)
candidates.head(5)
constituencies.head(5)

election_results.shape
candidates.shape
constituencies.shape

election_results.dtypes
candidates.dtypes
constituencies.dtypes

election_results.columns
candidates.columns
constituencies.columns
```

```python
# Task 3: Null Value Handling
election_results[election_results.isna().any(axis=1)]
election_results.isna().sum() #election_results.isnull()
candidates.isnull().sum()
constituencies.isnull().sum()

election_results['votes'] = election_results['votes'].fillna(election_results['votes'].mean())
election_results = election_results.dropna()
```

```python
# Task 4: Duplicate Handling
election_results[election_results.duplicated()]
election_results.drop_duplicates()

candidates[candidates.duplicated()]
candidates.drop_duplicates()

constituencies[constituencies.duplicated()]
constituencies.drop_duplicates()
```

```python
# Task 5: Merge Data
ec = pd.merge(election_results,candidates,on='candidate_id')
ecc = pd.merge(ec,constituencies,on='constituency_id')
data = ecc[['candidate_name','party','constituency_name','state','votes']]
data = data.drop_duplicates().dropna()
```

```python
# Task 6: Apply Function
def voting_category(vote):
    if vote > 80000:
        return 'High'
    elif vote > 70000:
        return 'Medium'
    else:
        return 'Low'
    
data['vote category'] = data['votes'].apply(voting_category)
```

```python
# Task 8: Aggregate Functions

data['votes'].sum()
data['votes'].mean()
data['votes'].max()
data['votes'].min()
data['candidate_name'].count()
```

```python
# Task 9: Grouping


data.groupby('state')['votes'].sum()
data.groupby('party')['votes'].mean()
data.groupby('state')['votes'].max()
data.groupby('party')['votes'].sum()
```

```python
# Task 10: Sorting

data.sort_values('votes')
data.sort_values('candidate_name')
```

```python
# Task 11: Advanced Analysis

data.loc[data.groupby('state')['votes'].idxmax()]
par = data.groupby('party',as_index=False)['votes'].sum()
winner = par.sort_values('votes',ascending=False).head(1)

d = data.loc[data.groupby('candidate_name')['votes'].idxmax()]
d.sort_values('votes',ascending=False).head(3)
d.sort_values('votes',).head(3)

par = data.groupby('constituency_name',as_index=False)['votes'].sum()
winner = par.sort_values('votes',ascending=False).head(1)
winner
```

```python
party_votes = data.groupby('party', as_index=False)['votes'].sum()
total_votes = party_votes['votes'].sum()
party_votes['vote_percentage'] = (party_votes['votes'] / total_votes) * 100

print(party_votes)
```

```python
party_avg = data.groupby('party', as_index=False)['votes'].mean()
party_avg.sort_values('votes', ascending=False).head(1)
```

```python
winners = data.loc[data.groupby('constituency_name')['votes'].idxmax()]
winners.groupby('party', as_index=False)['constituency_name'].count().rename(columns={'constituency_name': 'constituencies_won'})
```
